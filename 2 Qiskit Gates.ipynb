# Quantum Gates


```python
from qiskit import *
import math
```

```python
from qiskit.tools.visualization import plot_bloch_multivector
```

```python
circuit = QuantumCircuit(1,1) #circuit with 1 qubit and 1 bit
circuit.rx(math.pi/2,0) #apply x gate on qubit
simulator = Aer.get_backend('statevector_simulator')
result = execute(circuit, backend = simulator).result()
statevector = result.get_statevector()
print(statevector)
%matplotlib inline
circuit.draw(output = 'mpl')
```

People refer to operations on quantum states as rotations on the bloch sphere. The X gates has taken our state from 0 to 1. So any operation for a individual qubit can be represented as a point on the surface of this sphere.

```python
plot_bloch_multivector(statevector) #plot statevector on bloch sphere
```

We can run this quantum circuit, execute measurements and get results.

```python
circuit.measure([0], [0]) #measure qubit 0 and put into bit 0
backend = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend = backend, shots = 1024).result()
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```

For the matrix presentation of the circuit, we just change the statevector to unitary.

```python
circuit = QuantumCircuit(1,1) #circuit with 1 qubit and 1 bit
circuit.x(0) #apply x gate on qubit
simulator = Aer.get_backend('unitary_simulator')
result = execute(circuit, backend = simulator).result()
unitary = result.get_unitary()
print(unitary)
```
